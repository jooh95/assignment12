```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.linalg as lin
```

# Set size of the number and standard deviation.

```python
num     = 1001
std     = 5 
```

# Define the function.

```python
def fun(x):
 
	f = np.sin(x) * (1 / (1 + np.exp(-x))) 
	f = np.abs(x) * np.sin(x)

	return f
```

# Define the size of lambda.

```python
l_idx = np.arange(-3,6)
```

# Define a function for getting theta values and an error through least square algorithm.

```python
def least_square(X, Y, p, l):
    A = np.ones((X.shape[0], p))
    zero = np.zeros((p, 1))
    for i in range(0,p):
        A.T[i] = X ** i
    I = np.identity(p)
    I = (2 ** int(l)) * I
    n_A = np.concatenate((A, I), axis = 0)
    n_Y = np.concatenate((Y.reshape(-1),zero.reshape(-1)), axis = 0)
    
    n_p = np.dot(np.dot(lin.inv(np.dot(n_A.T, n_A)), n_A.T), n_Y)
    
    w_sum = 0
    for i in range(0,num):
        w_sum += ((np.dot(A,n_p)[i] - Y[i]) ** 2)
    
    l_sum = 0
    for i in range(0,num):
        l_sum += l * (A.reshape(-1)[i] ** 2)

    return np.dot(A,n_p), (w_sum - l_sum)
```

# Set values for clean and noisy points.

```python
n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x) 			# clean points
y2      = y1 + nn * std		# noisy points
```

# Visualize noisy, clean, least square curve, and energy

```python
for i in range(6,16):
    error_idx = np.zeros(0)
    for j in l_idx:
        n_y, error = least_square(x, y1, i, j)
        error_idx = np.append(error_idx, error)
        
        print("Paramter = ", i, "Lamda = ", 2 ** int(j))
        print("Red = Noisy Data, Green = Clean Data, Least Square Curve = Blue")
        plt.title("Noisy Data")
        plt.plot(x, y2, 'r')
        
        plt.title("Clean Data")
        plt.plot(x,y1, 'g')
        
        plt.title("Least Square Curve")
        plt.plot(x, n_y,'b')
        plt.show()
    print("Paramter = ", i, "Lamda = ", 2 ** int(j))
    plt.title("Energy Graph")
    plt.plot(l_idx, error_idx)
    plt.show()
```
